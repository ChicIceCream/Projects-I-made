```python
import librosa
import librosa.display
import IPython.display as ipd
import os
import numpy as np
```

```python
for file in os.listdir('audio data/'):
    print(file)
```

```python
ipd.Audio('audio data/OAF_back_fear.wav')
```

```python
def feature_extraction(file_path):
    # load the audio file
    x, sample_rate = librosa.load(file_path, res_type='kaiser_fast')
    # extract features from the audio
    mfcc = np.mean(librosa.feature.mfcc(y=x, sr=sample_rate, n_mfcc=50).T, axis=0)
    
    return mfcc
```

```python
features = {}
directory = 'audio data/'
for audio in os.listdir(directory):
    audio_path = directory+audio
    features[audio_path] = feature_extraction(audio_path)
```

```python
audio_path

features[audio_path], len(features[audio_path])
```

```python
for audio_path, feature in features.items():
    print(f"Features for {audio_path}: {feature}\n")
```

```python
import matplotlib.pyplot as plt

# Loop through all audio files' features and plot them
for audio_path, mfcc_features in features.items():
    plt.figure(figsize=(10, 4))
    plt.plot(mfcc_features)
    plt.title(f"MFCC Features for {audio_path}")
    plt.xlabel("MFCC Coefficients")
    plt.ylabel("Amplitude")
    plt.show()

```
